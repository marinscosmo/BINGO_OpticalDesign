```python
%matplotlib inline
import os,sys
import scipy 
import matplotlib
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from matplotlib import rc
from matplotlib import cm

import OpticalDesign as OD
import ModelDesign as MODEL

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
font = {'weight' : 'bold',
        'size'   : 22}
matplotlib.rc('font', **font)
```

```python
data    = OD.load_data("optimization")
recdata = OD.fit_parameters(X=data.x,Y=data.y)
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.2, wspace=.3)
fig.subplots_adjust(bottom=0.0, left=0.025, top = 2, right=2.)

#--------------------
ax1 = fig.add_subplot(321)
plt.plot(range(len(recdata.z-data.z)),recdata.z-data.z,color="red")
plt.axhline(y=0,c="black",ls="dashed")
ax1.tick_params(labelsize=20)
ax1.set_ylabel(r"$z_{\rm{fit}}-z$",fontsize=30)

ax2 = fig.add_subplot(322)
plt.plot(recdata.z,label=r"$z_{fit}$", color="red")
plt.plot(data.z,label=r"$z$", color="blue")
ax2.set_ylabel("$z (cm)$",fontsize=30)
ax2.tick_params(labelsize=20)
plt.legend(loc="best",fontsize=20)

#--------------------
ax3 = fig.add_subplot(323)
plt.plot(range(len(recdata.theta-data.theta)),recdata.theta-data.theta,color="red")
plt.axhline(y=0,c="black",ls="dashed")
ax3.tick_params(labelsize=20)
ax3.set_ylabel(r"$\theta_{fit}-\theta$",fontsize=30)

ax4 = fig.add_subplot(324)
plt.plot(recdata.theta,label=r"$\theta_{fit}$", color="red")
plt.plot(data.theta,label=r"$\theta$ ", color="blue")
ax4.set_ylabel(r"$\theta (deg)$",fontsize=30)
ax4.tick_params(labelsize=20)
plt.legend(loc="best",fontsize=20)

#--------------------
inds = np.where(data.phi==180)[0]
ax5 = fig.add_subplot(325)
plt.plot(range(len(recdata.phi-data.phi)),recdata.phi-data.phi, color="red")
plt.axhline(y=0,c="black",ls="dashed")
ax5.set_ylabel(r"$\phi_{fit}-\phi$",fontsize=30)
ax5.tick_params(labelsize=20)

ax6 = fig.add_subplot(326)
plt.plot(recdata.phi,label=r"$\phi_{fit}$", color="red")
plt.plot(data.phi   ,label=r"$\phi$", color="blue")
ax6.set_ylabel(r"$\phi (deg)$",fontsize=30)
ax6.tick_params(labelsize=20)
plt.legend(loc="best",fontsize=20)

plt.savefig("images/comparation_horns_pol.jpg", dpi=50, bbox_inches='tight');
```

```python
print("sigma(zfit-z)         = {}".format(np.std(recdata.z-data.z)))
print("sigma(thetafit-theta) = {}".format(np.std(recdata.theta-data.theta)))
print("sigma(phifit-phi)     = {}".format(np.std(recdata.phi-data.phi)))
```
